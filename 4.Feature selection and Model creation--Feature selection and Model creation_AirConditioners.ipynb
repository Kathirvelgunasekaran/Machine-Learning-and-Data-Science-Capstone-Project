```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
import pickle
```

```python
# Load dataset
dataset = pd.read_csv("PreprocessAirConditioners.csv")
```

```python
dataset
```

```python
# Input and output split
independent = dataset[['no_of_ratings', 'discount_price', 'actual_price']]  # Input features
dependent = dataset[["ratings"]]  # Output feature
```

```python
independent
```

```python
dependent
```

```python
# Split Trainset and Testset (70:30 ratio)

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
# Standardization of input features
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train #Now X_train variable have standardized values. using this value going to create Model
```

```python
X_test
```

```python
# Apply PCA for feature reduction
pca = PCA(n_components=2)  # Reduce to 2 principal components, change this as needed
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
# Standardize output features (optional, if needed for output)
scy = StandardScaler()
y_train = scy.fit_transform(y_train)
y_test = scy.transform(y_test)
```

```python
# Create Random Forest Regressor model
regressor = RandomForestRegressor(n_estimators=50, random_state=0)
regressor.fit(X_train_pca, y_train)
```

```python
# Make predictions on the test set
y_pred = regressor.predict(X_test_pca)
```

```python
# Evaluate model using R2 score
r_score = r2_score(y_test, y_pred)
print(f"R2 Score: {r_score}")
```

```python
r_score
```

```python
# Save the model using pickle
filename = "finalized_model_randomforest_with_pca.sav"
pickle.dump(regressor, open(filename, 'wb'))
```

```python
# Example of new user input for prediction
preinput = sc.transform([[1, 2, 5]])  # User-provided input
preinput_pca = pca.transform(preinput)  # Apply PCA to the new input
```

```python
# Load the saved model and make predictions
loaded_model = pickle.load(open(filename, 'rb'))
result = loaded_model.predict(preinput_pca)
```

```python
# Convert the predicted result back to original scale
preoutput = scy.inverse_transform([result])
print(f"Predicted Output: {preoutput}")
```
