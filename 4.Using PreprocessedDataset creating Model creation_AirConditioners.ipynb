```python
import pandas as pd
```

```python
dataset=pd.read_csv("PreprocessAirConditioners.csv") 
```

```python
dataset
```

```python
dataset.columns #It'll show all dataset columns. we can use it for further process related.
```

```python


#Input and output split
#Data was already preprocessed. Now we are going to perform Input and output variables split
independent=dataset[['no_of_ratings', 'discount_price', 'actual_price']] 
independent #Now 'R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York' column values seperately assigned into independent varaiable, Independent means input, Dependent means output.
```

```python
#Now 'Profit' column value seperately assigned into dependent varaiable, dependent means output.
dependent=dataset[["ratings"]] #Now Salary value seperately assigned into dependent varaiable,
dependent
#Now we got seperatley independent and dependent varaiables. now sucessfully splited input and output
```

```python
#Split Trainset and Testset. 
#Same code we are using for Simple linear regression
#Using sklearn library going to split Trainset and Testset. it's supporting for Machine Learning.
#respective folder(skleran) under this function(model_selection) we are going to (import) or calling train_test_split
#Not calling all functions under this respective folder(skleran) under this function(model_selection) we are going to call train_test_split.
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0) #No varaiable assigned here.here parallely split input and output based on 70:30 ratio.
#we are using _(underscore) instead of space into vraiables(X_train,X_test,y_train,y_test)
#X means input, y means output, X input based on trainset and testset(X_train,X_test) and also y was output based trainset and testset(y_train,y_test) 
#if you want to use sklearn library based on trainset and testset you have to split it like this four variables(X_train,X_test,y_train,y_test).
# here we are passing Actual input@output Trainset@Testset 70% data, remaining 30% Trainset@Testset output dataset exist into X_test,y_test, going to use it for Model testing purpose.
#Actual input/output 70% available in dataset(X_train,y_train), remaining 30% going to use for Model testing purpose(X_test,y_test)
#As of now compltetd Trainset and Testset split.
```

```python
#Standardiztion Technique input preprocess
#Now we are going try with standardization technique. we will see how it working with standardization technique. where we seperated Trainset and testset center we are going to try.
from sklearn.preprocessing import StandardScaler 
sc=StandardScaler() 
#calling StandardScaler and assign to sc variable and function created
X_train=sc.fit_transform(X_train) 
#X_train varaiable sc(now algorithm in sc varaiable) "fit_transform(X_train)" -to find means value and standarization vale calculate for X_train and give that value.
X_test=sc.transform(X_test) 
#X_test varaiable sc(now algorithm in sc varaiable) "transform(X_test)" -"while using 'fit' means value and standarization vale calculate for X_train "which we calculated for (X_train) same like that it'll use same means value and standarization calculate value for X_test and gie that value.

#*here we are preprocessing x AND keeping y as original value. preprocessed x and original y value using supervised learning, we passed preprocessed x value to original y value to learn.
#in this case we are preprocessing only X.
```

```python
X_train #Now X_train variable have standardized values. using this value going to create Model
```

```python
X_test
```

```python
#Standardiztion Technique output preprocess
#Now we are going try with standardization technique. we will see how it working with standardization technique. where we seperated Trainset and testset center we are going to try.
from sklearn.preprocessing import StandardScaler 
scy=StandardScaler() #calling StandardScaler and assign to sc variable and function created
y_train=scy.fit_transform(y_train) #X_train varaiable sc(now algorithm in sc varaiable) "fit_transform(X_train)" -to find means value and standarization vale calculate for X_train and give that value.
y_test=scy.transform(y_test) #X_test varaiable sc(now algorithm in sc varaiable) "transform(X_test)" -"while using 'fit' means value and standarization vale calculate for X_train "which we calculated for (X_train) same like that it'll use same means value and standarization calculate value for X_test and gie that value.

#*here we are preprocessing x AND keeping y as original value. preprocessed x and original y value using supervised learning, we passed preprocessed x value to original y value to learn.
#in this case we are preprocessing only X.
```

```python
y_train
```

```python
#Model Creation
#ML algorithm all are provided by sklearn, we are creting this Model using trainset and tesset varaiable. 
from sklearn.ensemble import RandomForestRegressor # While calling, this line loaded into memory. here RandomForestRegressor may be function or class. In this line RandomForestRegressor considered as class
#respective folder(skleran) under this function(ensemble) we are going to (import) or calling RandomForestRegressor
regressor=RandomForestRegressor(n_estimators=50,random_state=0) #Parameterized function we are assigning into regressor varaiable.it consider as class, all the function under this class assign to regressor.
regressor.fit(X_train,y_train) #fit function we are calling it using access operator, now passing splited trainset input and output.
#here just calling sklearn folder under linear_model we are calling RandomForestRegressor function only. due to this kind of calling memory consumption will be low
#what are all the functions RandomForestRegressor under this class loaded into regressor variable.
#class and functions are just we can call it as procedure.
#First two lines are procedure. while calling fit() after execution this line it'll create tree structure Model.now Model is ready.
#Third line when we call fit method and passing materials(X_train,y_train). this formula respective place substitute values. For this particular block have Learning phase going to happen here,
#First two lines are procedure, Alway model will create once fit() method got executed and that model will save regressor varaiable. Random forest model will create here. regressor word can be different in another program.
#It'll check fit() method and then pakathila create model. Model is called AI Brain.
#https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestRegressor.html
#Above page we have many RandomForestRegressor parameters, depends on our need change and execute algorithm
#Attribute we have all the needed things once model crreated what are all we are going to get.
#change n_estimators=50 percentage value and check accuracy %.
```

```python
#Evaluation metrix(R2 using)
y_pred=regressor.predict(X_test) #y-pred is varaiable, all are saved in regressor(formula,weight,bais), predict is application function.
#Evaluation matrix should happened on Testdataset .Anyone of the Testset values not included into Trandataset values. so that we are using it for crosss predition
#y-test is Actual data(y_test), X_test is brain predicted outptut.so that we are using it for crosss prediction
```

```python
from sklearn.metrics import r2_score #using this r2_score function going to find R2 value.
r_score=r2_score(y_test,y_pred) #r2_score assigned to r_score variable, now passing Actual output(y_test) and predicted output(y_pred)
#r2_score result nearly 1 means it's Good model, nearly to 0 means bad model
#Now it nearly to 1. our model is performing well. our model learned well while performing crosscheck. evaluation matrix cross checking our model and it'll confirm particular model is performing well or not. 
```
